## In this example, we will use the encoders from the scikit-learn library. Specifically, the LabelEncoder of creating an integer encoding of labels and the OneHotEncoder for creating a one hot encoding of integer encoded values.

```python
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(len(integer_encoded))
```

```python
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
print(len(onehot_encoded))
print(len(onehot_encoded[0]))
#this means that there are some input words that weren't encoded..
```

```python
data=[]
path='C:\\Users\\Lamia\\AppData\\Roaming\\nltk_data\\corpora\\inaugural'
path=os.path.join(path, '*.txt')
#path
files = glob(path)
#print (files)
for f in files:
    with open(f) as document:
        data.append(document.read())
        
```

```python
len(data) #58 text files
data[2]
```

Preprocessing, cleaning and stemming

```python

REPLACE_NO_SPACE = re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])")
REPLACE_WITH_SPACE = re.compile("(\n\s*)|(\-)|(\--)|(\/)")

def review_to_words(document):
    words = REPLACE_NO_SPACE.sub("", document.lower())
    words = REPLACE_WITH_SPACE.sub(" ", words)
    return words
```

```python
clean_docs = [review_to_words(document) for document in data]
clean_docs[1]
```

LEt's start stemming all docs


```python
porter = nltk.PorterStemmer()
porter.stem(clean_docs[1])
```

See? this isn't working since the stemmer considers the whole sentence as one word...

```python

def stem_corpus(corpus_list):
    Clean_stem_docs=[]    
    porter = nltk.PorterStemmer() #constructor.
    #tokenize each document words and put put each doc in stemmed_doc list then remove it to corpus list named Clean_stem_docs
    for i in range (len(corpus_list)):
        stemmed_doc=[]
        token_words=word_tokenize(corpus_list[i])
        for word in token_words:
            stemmed_doc.append(porter.stem(word))
        stemmed_doc=" ".join(stemmed_doc)
        Clean_stem_docs.append(stemmed_doc)
    return Clean_stem_docs


clean_stemmed_docs=stem_corpus(clean_docs)
```

```python
clean_stemmed_docs
#clean_stemmed_docs[1]
```

```python
vocabulary_size=5000
vectorizer = CountVectorizer(max_features=vocabulary_size) #constructor.
features = vectorizer.fit_transform(clean_stemmed_docs)
#features
#print(features[0])
#print(vectorizer.get_feature_names())
Arr_features=features.toarray()
#len(Arr_features[1]) #definitely=5000
print(Arr_features)
```

```python
for i in range(0,5000):
    if Arr_features[0][i]>1:
        print (Arr_features[0][i])
```
